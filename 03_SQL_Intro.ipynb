```python
# Part 3 - SQL intro

# Import necessary libraries
import sqlite3
import pandas as pd
```

```python
# Create a new database for this
conn = sqlite3.connect("superstore.db")
```

```python
# Writing the cleaned csv to the database
df = pd.read_csv("/Users/thomassimmons/Coding/SS/Superstore_cleaned.csv")

df.to_sql("superstore_sales", conn, if_exists="replace", index=False)
```

```python
pd.read_sql("PRAGMA table_info (superstore_sales);", conn)
```

```python
pd.read_sql("SELECT * FROM superstore_sales LIMIT 5;", conn)
```

```python
pd.read_sql("SELECT City FROM superstore_sales WHERE City = 'Los Angeles' LIMIT 5;", conn)
```

```python
# Closing connection 
conn.close()
```
