```python
# Part 4 - SQL analysis

import sqlite3
import pandas as pd

conn = sqlite3.connect("/Users/thomassimmons/Coding/SS/superstore.db")
```

```python
pd.read_sql("PRAGMA table_info (superstore_sales);", conn)
```

```python
pd.read_sql("SELECT * FROM superstore_sales LIMIT 5;", conn)
```

```python
# Sales in south Louisville kY
pd.read_sql("SELECT * FROM superstore_sales WHERE State = 'Kentucky' AND City = 'Louisville' AND Region = 'South';", conn)
```

```python
pd.read_sql("""
    SELECT Region, SUM(Sales) AS Total_Revenue
    FROM superstore_sales
    GROUP BY Region
    ORDER BY Total_Revenue DESC;
""", conn)
```

```python
pd.read_sql("""
            SELECT [Product Name], SUM(Sales) AS Total_REVENUE
            FROM superstore_sales
            GROUP BY [Product Name]
            ORDER BY Total_Revenue DESC
            LIMIT 5;
            """, conn)
```

```python
pd.read_sql("""
            SELECT strftime('%Y-%m', [Order Date]) AS Month, SUM(Sales) AS Monthly_Revenue
            FROM superstore_sales
            GROUP BY Month
            ORDER BY Month;
            """, conn)
```

```python
pd.read_sql("""
            SELECT [Customer ID], AVG(Sales) AS Avg_Basket_Value
            FROM superstore_sales
            GROUP BY [Customer ID]
            ORDER BY Avg_Basket_Value DESC;
            """,conn)
```
